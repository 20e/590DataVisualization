```python
chars = ["Boba Fett", "Yoda", "Jabba Desilijic Tiure", "Darth Vader", "Obi-Wan Kenobi",
         "Beru Whitesun lars", "Mon Mothma"]
```

```python
char = "Boba Fett"
```

```python
chars_pks = []
for char in char_input:
    if char['fields']['name'] in chars:
        chars_pks.append(char['pk'])
```

```python
chars_pks
```

```python
numbers = [0]
```

```python
import pandas as pd
```

```python
chars_list = pd.DataFrame(
    {'names': chars,
     'pk': chars_pks
    })
```

```python
chars_list['number'] = list(range(0, len(chars)))
```

```python
chars_list
```

```python
int(chars_list[chars_list['pk']==4]['number'])
```

```python
film_p = []
char_ps = []
for film in films:
    char_subset = film['fields']['characters']
    char_p = []
    for char in chars_pks:
        if char in char_subset:
            char_p.append(int(chars_list[chars_list['pk']==char]['number']))
            if film['fields']['title'] not in film_p:
                film_p.append(film['fields']['title'])
    char_ps.append(char_p)
```

```python
film_p
```

```python
char_ps
```

```python
chars_film = pd.DataFrame(
    {'films': film_p,
     'characters': char_ps
    })
```

```python
chars_film
```

```python
links = []
for j in range(chars_film.shape[0]):
    for i in range(len(chars_film['characters'])):
        for a in range(len(chars_film['characters'][j])-i):
            cell = {}
            cell['source']= chars_film['characters'][j][i]
            cell['target']= chars_film['characters'][j][i+a]
            cell['value'] = chars_film['films'][j]
            links.append(cell)
```

```python
links
```

```python
node_data = chars
link_data = links

graph = bqplot.Graph(node_data=node_data, link_data=link_data)

fig = bqplot.Figure(marks = [graph])
display(fig)
```
